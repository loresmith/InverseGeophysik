# Theorie

## Singulärwertzerlegung einer Matrix

Sei die rechteckige Matrix $\mathbf G \in \mathbb{R}^{m \times n}$ mit $m \ge n$ und $rg(\mathbf G) = r$ gegeben. Dann ergibt die Singulärwertzerlegung der Matrix $\mathbf G$ die beiden orthogonalen Matrizen $\mathbf U \in \mathbb{R}^{m \times m}$ und $\mathbf V \in \mathbb{R}^{n \times n}$, sowie die Matrix der Singulärwerte $\mathbf S$. 

\begin{equation}
\mathbf S = 
\begin{bmatrix}
\mathbf S_r & 0 \\
0 & 0
\end{bmatrix}
\in \mathbb{R}^{m \times n}
\label{eq:1} \tag{1}
\end{equation} 

mit 

\begin{equation}
\mathbf S_r = 
\begin{bmatrix}
\sigma_1 & 0 & 0 \\
0 & \ddots & 0 \\
0 & 0 & \sigma_r
\end{bmatrix}
\in \mathbb{R}^{r \times r}
\label{eq:2} \tag{2}
\end{equation} 

mit den Singulärwerten $\sigma_1 \ge \ldots \ge \sigma_r > 0$

Die Darstellung der Singulärwertzerlegung sieht dann folgendermaßen aus:

\begin{equation}
\mathbf G = \mathbf U \mathbf S \mathbf V^T
\label{eq:3} \tag{3}
\end{equation} 

## Abgeschnittene Singulärwertzerlegung

Zunächst wird die Matrix $\mathbf G$ in dem Vorwärtsproblem $\mathbf G \mathbf m = \mathbf d$ mit $\mathbf G_k$ für die nichtverschwindenden Singulärwerte $\sigma_1,\ldots, \sigma_k$ ersetzt. Damit bleiben in $\mathbf G$ die Singulärwerte bis $\sigma_k$ unverändert und die Singulärwerte $\sigma_{k+1}, \sigma_{k+2}, \ldots, \sigma_r$ werden gleich Null gesetzt. 

\begin{equation}
\mathbf G_k = \mathbf U_k \mathbf S_k \mathbf V_k^T
\label{eq:4} \tag{4}
\end{equation} 

Nach der Multiplikation der Gleichung mit $\mathbf U_k^T$, $\mathbf S_k^{-1}$ und $\mathbf V_k$ kann das Modell $\mathbf m$ mit den Daten $\mathbf d$ aus dem Vorwärtsproblem berechnet werden.

\begin{equation}
\mathbf m = \mathbf V_k \sigma_k^{-1} \mathbf U_k^T \mathbf d
\label{eq:5} \tag{5}
\end{equation} 

Die verallgemeinerte Inverse wird durch $\mathbf G^{\dagger} = \mathbf V_k \sigma_k^{-1} \mathbf U_k^T$ (Moore-Penrose-Pseudoinverse) beschrieben. In der Summenschreibweise sieht dies folgendermaßen aus: 

\begin{equation}
\mathbf m = \sum_{i=1}^k \frac{1}{\sigma_i} V_i~ \mathbf d~ U_i^T 
\label{eq:6} \tag{6}
\end{equation} 

oder 

\begin{equation}
\mathbf m = \sum_{i} \frac{f_i}{\sigma_i} V_i~ \mathbf d~ U_i^T 
\label{eq:7} \tag{7}
\end{equation} 

mit $f_i = 1$

## Wahl des Pseudorangs

Bei der Inversion synthetischer Modelle wird der Pseudorang für die Singulärwertzerlegung nach dem Diskrepanzprinzip gewählt. Hierfür wird das Model $\mathbf m$ für die Ränge $1, \ldots, r$ mithilfe der abgeschnittenen Singulärwertzerlegung und $f_1 = f_2 = \ldots = f_r = 1$ berechnet. Anschließend erfolgt die Berechnung des Residuals mit $\mathbf r = \mathbf G \mathbf m - \mathbf d$ für jedes Modell $m_i$. Das Residual wird dann in der Berechnungsformel des chi-Quadrat-Werts angewendet. 

\begin{equation}
\chi^2 = \frac{1}{N} \sum_{i=1}^r \left( \frac{\mathbf d - \mathbf G m_i}{\boldsymbol\epsilon} \right)^2
\label{eq:8} \tag{8}
\end{equation} 

Dabei beschreibt $\boldsymbol\epsilon$ den Vektor des Fehlermodells und $N$ die Länge des Datenvektors $\mathbf d$.
Ein chi-Quadrat von 1, welches gleichbedeutend mit der Residualnorm von $\sqrt{N}$ ist, zeigt, dass die invertierten Daten auch noch mit Rauschanteilen eine gute Anpassung liefern. Somit wird das $\chi^2$ auch durch die Residualnorm mit verrauschten Daten $\left|\left|~\mathbf{\hat{r}}~\right|\right|^2_2 = \left|\left|~\mathbf{\hat{d}}-\mathbf{\hat{G}} \mathbf m~\right|\right|^2_2$ berechnet:

\begin{equation}
\boldsymbol\chi^2 = \frac{1}{N} \left|\left|~\mathbf{\hat{r}}~\right|\right|^2_2
\label{eq:9} \tag{9}
\end{equation}

Der optimale Pseudorang $p$ für die abgeschnittene Singulärwertzerlegung ist nach der Anwendung des Kriteriums $min = \boldsymbol\chi^2 -1$ auf den entstandenen Vektor $\boldsymbol\chi^2$ gefunden, indem $min$ den $p$ten-Wert in dem Vektor kennzeichnet. 

Zusätzlich kann die Residualnorm $\left|\left|~\mathbf{\hat{r}}~\right|\right|^2_2$ mit der Modellnorm $\left|\left|~\mathbf m~\right|\right|^2_2$ gegen die Anzahl der Singulärwerte geplottet werden. \
In diesem Plot ist ergibt dann der Schnittpunkt der Residualnorm $\left|\left|~\mathbf{\hat{r}}~\right|\right|^2_2$ mit der Residualnorm $\sqrt{N}$ den Pseudorang (siehe Unterkapitel [ChiSquare](#ChiSquare)). Weiterhin kann auch die Modellnorm gegen die Residualnorm $\left|\left|~\mathbf{\hat{r}}~\right|\right|^2_2$ abgebildet werden. Dieses Diagramm wird aufgrund ihrer Form die L-Kurve genannt und besitzt je einen steilen und einen flachen Teil, sowie  einen Übergangspunkt zwischen den beiden Abschnitten. Der Pseudorang befindet sich in einem Abschnitt von konstanten Werten (ein sogenanntes Plateau) in der L-Kurve.


## Inversion nach Occam Prinzip (Smoothness Constraints)

Nach dem Prinzip von Occam soll aus allen möglichen Lösungen immer die einfachste Lösung ausgewählt werden. Für die Inversion mit Glattheits-Nebenbedingungen wird ein Regularisierungsparameter $\lambda$ bestimmt, der die Bedingung von Occam am besten erfüllt. Der Regularisierungsparameter $\lambda$, auch Wichtungsfaktor genannt, muss daher so gewählt werden, dass die Komplexität und die Merkmale des Modells erhalten bleiben, jedoch durch eine zu starke Glättung nicht unkenntlich gemacht werden. Ein zu niedriger Regularisierungsparameter kann hingegen zu kleine Merkmale oder Artefakte im Modell darstellen. Nach Occam wird der $\lambda$-Wert in der Inversion genutzt, der möglichst geringe Kontraste im Modell erzeugt und eine Überglättung des Modells vermeidet. \
Das Ziel der Inversion ist dann die Minimierung der gewichteten Summe (Residuum + Constraints):

\begin{equation}
\left|\left|~\mathbf G \mathbf m - \mathbf d~\right|\right|^2_2 + \lambda^2 \left|\left|~\mathbf W \mathbf m~\right|\right|^2_2 \rightarrow min
\label{eq:10} \tag{10}
\end{equation}

Der Term $\lambda^2 \left|\left|~\mathbf W \mathbf m~\right|\right|^2_2$ ist der Regularisierungsterm, der die Rauigkeit des Modells minimiert. Die Matrix $\mathbf W \in \mathbb{R}^{2m \times n}$ regularisiert das Modell in die $x$- und $y$-Richtung.

\begin{equation}
\mathbf W = 
\begin{bmatrix}
\mathbf W_x \\
\mathbf W_y 
\end{bmatrix} =
\begin{bmatrix}
-1 & 1 & 0 & \dots & \dots \\
0 & -1 & 1 & 0 & \dots & \\
\vdots & \vdots & \ddots & \ddots & \vdots & \\
\dots & \dots & 0 & -1 & 1 & \\
-1 & 0 & \dots & 1 & 0 & \vdots \\
\vdots & -1 & \dots & 0  & 1 &
\end{bmatrix}
\label{eq:11} \tag{11}
\end{equation} 

Damit kann die Minimierung für eine unabhängige Regularisierung nach $x$ und $y$ auch folgendermaßen ausgedrückt werden:

\begin{equation}
\left|\left|~\mathbf G \mathbf m - \mathbf d~\right|\right|^2_2 + \lambda_x \left|\left|~\mathbf W_x \mathbf m~\right|\right| + \lambda_y \left|\left|~\mathbf W_y \mathbf m~\right|\right| \rightarrow min
\label{eq:12} \tag{12}
\end{equation}


Die Normalengleichung zu dem Minimierungsproblem in $\ref{eq:10}$ lautet:

\begin{equation}
\left( \mathbf G^T \mathbf G + \lambda^2 \mathbf W^T \mathbf W \right) \mathbf m = \mathbf G^T \mathbf d
\label{eq:13} \tag{13}
\end{equation}

Daraus ergibt sich dann die Berechnung des Modells $\mathbf m$:

\begin{equation}
\mathbf m = \left( \mathbf G^T \mathbf G + \lambda^2 \mathbf W^T \mathbf W \right)^{-1} \mathbf G^T \mathbf d
\label{eq:14} \tag{14}
\end{equation}

Das optimale $\lambda$ wird wie bei der Wahl des Pseudorangs durch das $\chi^2$-Kriterium (Diskrepanzprinzip) bestimmt. Dafür wird ein $\boldsymbol\lambda$-Vektor mit einer logarthmischen Größenordnung angenommen und in der Inversion nach Occam angewendet. Das Minimum des $\chi^2$-Kriterium zeigt den Wert im $\boldsymbol\lambda$-Vektor an, der am besten zur Regularisierung des Modells geeignet ist ([Occam](#Occam)).

## Modellauflösungsmatrix und Informationsdichtematrix

### Abgeschnittene Singulärwertzerlegung

Die Modellauflösungsmatrix wird wie folgt berechnet:

\begin{equation}
\mathbf R^M = \mathbf G^{\dagger} \mathbf G = \mathbf V_k \sigma_k^{-1} \mathbf U_k^T \mathbf U_k \sigma_k \mathbf V_k^T = \mathbf V_k \mathbf V_k^T
\label{eq:15} \tag{15}
\end{equation}

Die Informationsdichtematrix ergibt sich aus:

\begin{equation}
\mathbf R^D = \mathbf G \mathbf G^{\dagger} = \mathbf U_k \sigma_k \mathbf V_k^T \mathbf V_k \sigma_k^{-1} \mathbf U_k^T = \mathbf U_k \mathbf U_k^T
\label{eq:16} \tag{16}
\end{equation}

### Invesion nach Occam

Die Modellauflösungsmatrix für die regularisierte Inversion wird durch folgende Gleichung beschrieben:

\begin{equation}
\mathbf R^M = \mathbf G^{\dagger} \mathbf G = \left( \mathbf G^T \mathbf G + \lambda^2 \mathbf W^T \mathbf W \right)^{-1} \mathbf G^T \mathbf G
\label{eq:17} \tag{17}
\end{equation}

mit der generalisierten Inversen $\mathbf G^{\dagger} = \left( \mathbf G^T \mathbf G + \lambda^2 \mathbf W^T \mathbf W \right)^{-1} \mathbf G^T$

Die Informationsdichtematrix der regularisierten Inversion liefert die folgende Beziehung:

\begin{equation}
\mathbf R^D = \mathbf G \mathbf G^{\dagger} = \mathbf G \left( \mathbf G^T \mathbf G + \lambda^2 \mathbf W^T \mathbf W \right)^{-1} \mathbf G^T
\label{eq:18} \tag{18}
\end{equation}

**Modelauflösungsmatrix**

Die Matrix gibt die Auflösung der Zellen in den Modellspalten an. Eine Anomalie wird dabei unterschiedlich gut in den jeweiligen Zellen aufgelöst. Bei der Betrachtung von nur einer Zelle im Modell wird die Modellauflösung $\mathbf R^M_{ij}$ besonders gut deutlich ([Auflösungsmatrizen](#Auflösungsmatrizen)). Der Index $j$ beschreibt dabei die $j$-te Zelle im Modell. Von dieser Zelle ausgehend, kann in einem Plot der Modellauflösungsmatrix beobachtet werden, wie stark sich eine Anomalie in der $j$-ten Zelle auf die benachbarten Zelle $i$ auswirkt. Das heißt, wie viel von der Anomalie wird auf die Zelle $i$ transformiert. Eine perfekte Auflösung wird durch $\mathbf R^M = \mathbf I$ (Einheitsmatrix) erreicht, d.h. es entstehen keine Artefakte in den benachbarten Zellen und das die Anomalie nur in der $j$-ten Zelle aufgelöst wird. Ein optimaler Wert für die Modellauflösung ist der Wert Eins und alle niedrigeren Werte verursachen eine geringere Auflösung im Modell, sodass die Anomalie nicht realitätsgetreu dargestellt wird.

**Informationsdichtematrix**

Diese Matrix beschreibt hingegen die Auflösung der Modellzeilen. Das invertierte Modell zeigt hier auch eine perfkete Auflösung, wenn es mit dem synthetischen Modell übereinstimmt und $\mathbf R^D = \mathbf I$ gilt. Der Wert Eins entspricht dabei einem Deltaimpuls, der nur in einer Zelle der Modellzeile vorkommt, d.h. diese eine Zelle wird perfekt aufgelöst. Treten jedoch Nebendiagonalelememte in der Informationsdichtematri auf, dann liegt keine optimale Auflösung des Modells mehr vor. Die Nebendiagonalelememte müssen dann gelättet werden.

```julia
#function chiSquare(sv, i, pMax, d, G, W, model, error)
function chiSquare(sv, i, pMax, d, G, model, error)
    #pMax = rank(G)
    

#chi, modelNorm, residualNorm,  = chiSquare(pMax, d, G, error)

#f = zeros(length(sv.S))
#modelNorm = zeros(pMax)
#residualNorm = zeros(pMax)
#residual = zeros(pMax)
#chi = zeros(pMax)
    
    #UTd = transpose(sD.U) * (d ./ error)
    
#for p in 1:pMax
    #f[p] = 1.0  # increasingly set f to one
    #model = sv.V * (f ./ sv.S .* UTd)
    #model = TSVD(sv, p, d, error)
    #modelNorm[i] = norm(W * model)
    modelNorm[i] = norm(model)
    residual = (G * model - d) ./ error
    residualNorm[i] = norm(residual)
    chi[i] = sum(residual.^2) / length(d)
#end
        
    #return chi, residualNorm, modelNorm, residual
    return chi[i], modelNorm[i], residualNorm[i], residual[i]
end
```

## Berechnung der Ableitungsmatrix

```julia
function DevMatrix(yweight, nx, ny)
    
    Wy = zeros(nx*(ny-1), nx*ny)
    Wx = zeros(nx*(ny-1), nx*ny)
    
    for i = 1:nx*(ny-1)
        Wy[i, i] = -1
        Wy[i, i+nx] = +1
    end
    
    Wy = Wy .* yweight
    
    
    ind = 0
    for iy = 1:ny
        for ix = 1:nx-1
            ind += 1  
            i = (iy-1)*nx + ix
            Wx[ind, i] = -1
            Wx[ind, i+1] = +1
        end
    end
    
    W = vcat(Wx, Wy)
       
    return W
    
end
```

## Modellberechnung mit abgeschnittener Singulärwertzerlegung (TSVD)

```julia
function TSVD(sv, p, d, err)
    
    f = zeros(length(sv.S))
    #slow = zeros(10*11)
    
    #f = zeros(d,1)
    UTd = transpose(sv.U) * (d ./ err)
    
    #for i in 1:p
        f[1:p] .= 1.0
        slow = sv.V * (f ./ sv.S .* UTd)
    #end
    
    return slow
    
end 
```

## Modellberechnung durch Occam-Inversion

```julia
function Occam(G, W, lambda, d, error)
    
    GT = transpose(G)
    GTd = GT * (d ./ error)
    WTWl = transpose(W) * W .* lambda^2
    model = pinv(GT * G + WTWl) * GTd 
    
    return model
    
end 
```

## Berechnung der Geschwindigkeiten mithilfe der Zeitmittel-Gleichung

```julia
function WyllieModel(v_a, v_w, v_m, por, sat, nx, ny)
    
        
    #slow1 = (1 - por[1]) * (1 / v_m) + (por[1] * sat) * (1 / v_w) + por[1] * (1 - sat)* (1 / v_a)
    #slow2 = (1 - por[2]) * (1 / v_m) + (por[2] * sat) * (1 / v_w) + por[2] * (1 - sat) * (1 / v_a)
    #slow3 = (1 - por[3]) * (1 / v_m) + (por[3] * sat) * (1 / v_w) + por[3] * (1 - sat) * (1 / v_a)
    
    slow1 = (1 - por[1]) * (1 / v_m) + (por[1] * sat[1]) * (1 / v_w) + por[1] * (1 - sat[1])* (1 / v_a)
    slow2 = (1 - por[2]) * (1 / v_m) + (por[2] * sat[2]) * (1 / v_w) + por[2] * (1 - sat[2])* (1 / v_a)
    slow3 = (1 - por[3]) * (1 / v_m) + (por[3] * sat[3]) * (1 / v_w) + por[3] * (1 - sat[3])* (1 / v_a)
    slow4 = (1 - por[4]) * (1 / v_m) + (por[4] * sat[4]) * (1 / v_w) + por[4] * (1 - sat[4])* (1 / v_a)
    slow5 = (1 - por[5]) * (1 / v_m) + (por[5] * sat[5]) * (1 / v_w) + por[5] * (1 - sat[5])* (1 / v_a)
    
    slowsyn = zeros(nx, ny)
     
    slowsyn[1:10,1:10] .= slow1
    slowsyn[1:nx,11:20] .= slow2
    slowsyn[11:nx,1:10] .= slow3
    slowsyn[21:nx,21:ny] .= slow4
    slowsyn[1:20,21:ny] .= slow5
          
    return slowsyn
end
```
